```python
from tqdm import tqdm
```

```python
import random
from datetime import datetime, timedelta
from collections import defaultdict

# Initialize variables
morning_only = ['CHEF1', 'COZ1', 'COZ2', 'ASG1', 'ASG2', 'AUX1', 'CONF1', 'PIZ1', 'GARD1', 'MASSA1']
evening_only = ['CHEF2', 'COZ3', 'COZ4', 'ASG3', 'ASG4', 'AUX2', 'AUXPIZ1', 'PIZ2', 'GARD2', 'MASSA2']
roles = ['CHEF', 'COZ', 'ASG', 'AUX', 'CONF', 'PIZ', 'GARD', 'MASSA']
shifts = ['Morning', 'Evening']
days_in_september = [datetime(2023, 9, i+1) for i in range(30)]

employee_pool = {
    'CHEF': ['CHEF1', 'CHEF2'],
    'COZ': ['COZ1', 'COZ2', 'COZ3', 'COZ4'],
    'ASG': ['ASG1', 'ASG2', 'ASG3', 'ASG4'],
    'AUX': ['AUX1', 'AUX2', 'AUXPIZ1'],
    'CONF': ['CONF1'],
    'PIZ': ['PIZ1', 'PIZ2', 'AUXPIZ1'],
    'GARD': ['GARD1', 'GARD2'],
    'MASSA': ['MASSA1', 'MASSA2']
}

morning_only_pool = {role: [emp for emp in emps if emp in morning_only] for role, emps in employee_pool.items()}
evening_only_pool = {role: [emp for emp in emps if emp in evening_only] for role, emps in employee_pool.items()}
asg_pool = employee_pool.get('ASG', [])

# Initialize population
def initialize_population(population_size):
    population = []
    for _ in range(population_size):
        chromosome = []
        for day in days_in_september:
            for shift in shifts:
                for role in roles:
                    # Handle interchangeable roles: CONF1 and AUXPIZ1
                    if role == 'CONF':
                        interchangeable_pool = morning_only_pool.get('CONF', []) + morning_only_pool.get('AUXPIZ1', [])
                    elif role == 'AUX':
                        interchangeable_pool = evening_only_pool.get('AUX', []) + evening_only_pool.get('AUXPIZ1', [])
                    else:
                        interchangeable_pool = []

                    chosen_pool = []
                    if role == "ASG":
                        chosen_pool = asg_pool  # Choosing from ASG-specific pool
                    elif shift == "Morning":
                        chosen_pool = morning_only_pool.get(role, [])
                    elif shift == "Evening":
                        chosen_pool = evening_only_pool.get(role, [])
                    else:
                        chosen_pool = employee_pool.get(role, [])
                    
                    # Override with interchangeable pool if applicable
                    if interchangeable_pool:
                        chosen_pool = interchangeable_pool

                    if not chosen_pool:
                        raise IndexError(f"Empty employee pool for role {role} during {shift} shift.")
                        
                    employee = random.choice(chosen_pool)
                    chromosome.append((day, shift, role, employee))
        population.append(chromosome)
    return population


# Evaluate chromosome
def evaluate_chromosome(chromosome):
    score = 0
    
    # Initialize counters
    work_days = defaultdict(int)  # count of shifts per employee
    asg_separate = defaultdict(list)  # list of days ASG employees are working
    double_shift_days = defaultdict(list)  # list of days an employee is working a double shift
    sunday_off = set()  # employees who have a Sunday off
    monday_off = set()  # employees who have a Monday off
    
    # Iterate through chromosome to evaluate each assignment
    for day, shift, role, employee in chromosome:
        
        # Constraint: Morning/Evening only employees
        if employee in morning_only and shift == 'Evening':
            score -= 10
        if employee in evening_only and shift == 'Morning':
            score -= 10
        
        # Update work_days
        work_days[employee] += 1

        # Update ASG counter for separate workdays
        if role == 'ASG':
            asg_separate[employee].append(day)

        # Constraint: Sunday evening reduced staff
        if day.weekday() == 6 and shift == 'Evening':
            if role not in ['CHEF', 'PIZ']:
                score -= 10
                
        # Update days for double shifts
        if role != 'ASG':
            double_shift_days[employee].append(day)

    # Additional checks based on counted days

    # Constraint: 2 days off per week and 1 Sunday off per month for each employee
    for employee, count in work_days.items():
        if count > (30 - 4):  # more than (30 days - 4 days off)
            score -= (count - 26) * 5
        if employee not in sunday_off:
            score -= 10

    # Constraint: ASGs don't work together on the same day
    for days in asg_separate.values():
        if len(set(days)) < len(days):
            score -= 20

    # Constraint: Double shift only if next day is off
    for employee, days in double_shift_days.items():
        for day in days:
            next_day = day + timedelta(days=1)
            if next_day not in days:
                score -= 5
                
    # Soft Constraint: If Sunday off, Monday off as well
    for employee in sunday_off:
        if employee not in monday_off:
            score -= 1  # It's a soft constraint so penalty is low

    return score

# Tournament selection
def tournament_selection(population, scores, tournament_size=5):
    # Randomly select 'tournament_size' individuals and get the one with the highest fitness
    selected = random.choices(population, k=tournament_size)
    selected_scores = [scores[population.index(ind)] for ind in selected]
    return selected[selected_scores.index(max(selected_scores))]

def one_point_crossover(parent1, parent2):
    # Randomly select crossover point
    crossover_point = random.randint(1, len(parent1) - 1)

    # Create offspring by swapping tails
    offspring1 = parent1[:crossover_point] + parent2[crossover_point:]
    offspring2 = parent2[:crossover_point] + parent1[crossover_point:]

    return offspring1, offspring2

def mutate(chromosome, mutation_rate):
    for i in range(len(chromosome)):
        if random.random() < mutation_rate:
            day, shift, role, _ = chromosome[i]
            new_employee = random.choice(morning_only) if shift == 'Morning' else random.choice(evening_only)
            chromosome[i] = (day, shift, role, new_employee)

# Main GA loop
population_size = 1000
num_generations = 1000
crossover_rate = 0.8
mutation_rate = 0.2

population = initialize_population(population_size)
scores = [evaluate_chromosome(chromosome) for chromosome in population]

for generation in tqdm(range(num_generations)):
    new_population = []
    while len(new_population) < len(population):
        parent1 = tournament_selection(population, scores)
        parent2 = tournament_selection(population, scores)
        
        if random.random() < crossover_rate:
            offspring1, offspring2 = one_point_crossover(parent1, parent2)
        else:
            offspring1, offspring2 = parent1, parent2

        mutate(offspring1, mutation_rate)
        mutate(offspring2, mutation_rate)

        new_population.append(offspring1)
        new_population.append(offspring2)
    
    scores = [evaluate_chromosome(chromosome) for chromosome in new_population]
    population = new_population

# Best schedule found
best_schedule = population[scores.index(max(scores))]
```
